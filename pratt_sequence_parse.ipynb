```python
import re
import string
from collections import defaultdict
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import numpy as np

# ---------------------------------------------
# Global data structures and variables
# ---------------------------------------------
pattern_index_dicts = {}
flank1_index_dicts = {}
flank2_index_dicts = {}
dict_list = []
flank1_seq_list = []

# ---------------------------------------------
# Function definitions
# ---------------------------------------------
def create_dictionaries(index_list: list, dicts: dict):
  if len(index_list) == 0:
    print('** No wildcard found in pattern %s **' % pattern)
  else:
    for index in index_list:
      dicts[index] = defaultdict(int)

def count_aa(sequence: string, index_list: list, dicts: dict):
  for (i, x) in enumerate(sequence):
    if i in index_list:
      if i not in dicts:
        dicts[i] = defaultdict(int)
      if x != 'o' and x != 'x':
        x = x.upper()
      dicts[i][x] += 1

def process_flank1(seq_list: list):
  max_len_seq = max(seq_list, key = len)
  max_len = len(max_len_seq)
  index_list = [i for i in range(len(max_len_seq))]
  create_dictionaries(index_list, flank1_index_dicts)
  dict_list.insert(0, flank1_index_dicts)
  for seq in seq_list:
    if len(seq) < max_len:
      seq = seq.rjust(len(seq) + (max_len - len(seq)), 'o')  # Pad seq to left with 'o'
    count_aa(seq, index_list, flank1_index_dicts)

def analyze_freq(index_dicts: dict):
  print('-' * 70)
  # Construct DataFrame from dictionary
  df = pd.DataFrame(index_dicts).T
  # Label DataFrame table
  df.index.names = ['Position']
  df.columns.names = ['Amino Acid']
  pd.set_option('display.max_columns', None)

  print('Frequencies:')
  display(df)
  # Construct bar plot for Frequencies
  colors = plt.cm.tab20(np.linspace(0, 1, 20))
  df.plot(xlabel = 'Position',
          ylabel = 'Frequency',
          kind = 'bar',
          width = 1,
          color = colors,
          stacked = False,
          title = 'Amino Acid Frequency at Variable Positions')
  plt.grid(axis = 'y')
  # Position legend outside of plot
  plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left', borderaxespad = 0)
  plt.show()

  # # Calculate probabilty
  # df = df / occurrences
  # print('Probabilities:')
  # display(df)
  # # Construct bar plot for probabilities
  # colors = plt.cm.tab20(np.linspace(0, 1, 20))
  # df.plot(xlabel = 'Position',
  #         ylabel = 'Probability',
  #         kind = 'bar',
  #         width = 1,
  #         color = colors,
  #         stacked = False,
  #         title = 'Amino Acid Probability at Variable Positions')
  # plt.grid(axis = 'y')
  # # Position legend outside of plot
  # plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left', borderaxespad = 0)
  # plt.show()

# ---------------------------------------------
# Script starts here
# ---------------------------------------------
begin_parse = False
with open('P4_3A.txt.135.pat', 'r') as f:
  while True:
    line = f.readline().strip()
    # Stop reading file when this line is reached
    if line == 'PATTERN MATCHES:':
      break

    if begin_parse:
      tokens = line.split()
      if len(tokens) >= 5:
        # Process pattern
        if len(tokens[0]) == 1:  # ['A', '1:', '29.1904', '3(', '3)', 'G-D-C-Q-N-P-S']
          occurrences = int(tokens[3].translate(str.maketrans('', '', string.punctuation)))
          pattern = tokens[5]
          print('-' * 70)
          print('{} Occurrences: {} Pattern: {}'.format(tokens[0], occurrences, pattern))
          pattern_aa_list = re.split(r'-', pattern)
          # Find variable regions in pattern
          pattern_index_list = [i for (i, x) in enumerate(pattern_aa_list) if x == 'x' or x.startswith('[')]
          create_dictionaries(pattern_index_list, pattern_index_dicts)
          dict_list.append(pattern_index_dicts)
        elif len(tokens) == 6:
          if tokens[4][:1].isupper():
            pattern_seq = tokens[4]
            flank2 = tokens[5]
            flank2_index_list = [i for i in range(len(flank2))]
            if not flank2_index_dicts:
              create_dictionaries(flank2_index_list, flank2_index_dicts)
              dict_list.append(flank2_index_dicts)
              count_aa(pattern_seq, pattern_index_list, pattern_index_dicts)
              count_aa(flank2, flank2_index_list, flank2_index_dicts)
            else:
              count_aa(pattern_seq, pattern_index_list, pattern_index_dicts)
              count_aa(flank2, flank2_index_list, flank2_index_dicts)
          else:
            flank1 = tokens[4]
            flank1_seq_list.append(flank1)
            pattern_seq = tokens[5]
            count_aa(pattern_seq, pattern_index_list, pattern_index_dicts)
        elif len(tokens) == 7:
          flank1 = tokens[4]
          flank1_seq_list.append(flank1)
          pattern_seq = tokens[5]
          flank2 = tokens[6]
          flank2_index_list = [i for i in range(len(flank2))]
          if not flank2_index_dicts:
            create_dictionaries(flank2_index_list, flank2_index_dicts)
            dict_list.append(flank2_index_dicts)
            count_aa(pattern_seq, pattern_index_list, pattern_index_dicts)
            count_aa(flank2, flank2_index_list, flank2_index_dicts)
          else:
            count_aa(pattern_seq, pattern_index_list, pattern_index_dicts)
            count_aa(flank2, flank2_index_list, flank2_index_dicts)
      # Reached end of a pattern block
      elif len(tokens) == 0:
        if flank1_seq_list:
          process_flank1(flank1_seq_list)
        # Begin frequency analysis for pattern block
        for dic in dict_list:
          if dic:
           analyze_freq(dic)
        print()
        print('=' * 70)
        print()
        # Clear data structures for next pattern block
        pattern_index_dicts.clear()
        flank1_index_dicts.clear()
        flank1_seq_list.clear()
        flank2_index_dicts.clear()
        dict_list.clear()

    # Start processing lines after this line is reached    
    if line == 'Best patterns with alignments:':
      f.readline()
      begin_parse = True
```
