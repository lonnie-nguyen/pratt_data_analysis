```python
import re
import string
from collections import defaultdict
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

# ---------------------------------------------
# Global data structures and variables
# ---------------------------------------------
aa_list = []
index_list = []
index_dicts = {}
occurrences = 0
pattern = ''

# ---------------------------------------------
# Function definitions
# ---------------------------------------------
def create_dictionaries(index_list: list):
  if len(index_list) == 0:
    print('No wildcard found in pattern %s' % pattern)
  else:
    for index in index_list:
      index_dicts[index] = defaultdict(int)
  
def count_aa(line: string):
  words = re.split(r'\W+', line)
  for word in words:
    if len(word) == len(aa_list) and word.startswith(aa_list[0]):
      for (i, x) in enumerate(word):
        if i in index_list:
          # Increment frequency of amino acid
          index_dicts[i][x.upper()] += 1

def analyze_freq(index_dicts: dict):
  # Construct DataFrame from dictionary
  df = pd.DataFrame(index_dicts).T
  # Label DataFrame table
  df.index.names = ['Position']
  df.columns.names = ['Amino Acid']
  print('Frequencies:')
  print(df)
  print('----------------------------------------------------------------------')
  # Calculate probabilty
  df = df / occurrences
  print('Probabilities:')
  print(df)
  # Construct bar plot for probabilities
  df.plot(xlabel = 'Position',
         ylabel = 'Probability',
         kind = 'bar',
         stacked = False,
         title = 'Amino Acid Probability at Variable Positions')
  plt.grid(axis = 'y')
  # Position legend outside of plot
  plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left', borderaxespad = 0)
  plt.show()

# ---------------------------------------------
# Script starts here
# ---------------------------------------------
begin_parse = False
# Open file to parse
with open('P4_3A.txt.135.pat', 'r') as f:
  while True:
    line = f.readline().strip()
    # Stop reading file when this line is reached
    if line == 'PATTERN MATCHES:':
      break

    if begin_parse:
      tokens = line.split()
      if len(tokens) >= 5:
        if len(tokens[0]) == 1:
          occurrences = int(tokens[3].translate(str.maketrans('', '', string.punctuation)))
          pattern = tokens[5]
          print('{} Occurrences: {} Pattern: {}'.format(tokens[0], occurrences, pattern))
          aa_list = re.split(r'-', pattern)
          index_list = [i for (i, x) in enumerate(aa_list) if x == 'x' or x.startswith('[')]
          create_dictionaries(index_list)
        elif len(index_list) != 0:
          count_aa(line)
      # Reached end of a pattern block
      elif len(tokens) == 0:
        # Begin frequency analysis for pattern block
        analyze_freq(index_dicts) if len(index_list) != 0 else print(index_dicts)
        print()
        # Clear data structures for next pattern block
        aa_list.clear()
        index_list.clear()
        index_dicts.clear()

    # Start processing lines after this line is reached
    if line == 'Best patterns with alignments:':
      f.readline()  #Skip line: '         fitness     hits(seqs)    Pattern'
      begin_parse = True
```
